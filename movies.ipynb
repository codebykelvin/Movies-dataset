```python
# lets bring in our friends to kickstart the party:) 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusting the configuration of the graph 

pd.options.mode.chained_assignment = None

```

```python
# next lets bring in our starters 
df= pd.read_csv('/Users/kelvincheng/Downloads/movies.csv')
```

```python
#checking what's inside our starters

df.head()
```

```python
#checking if starters is complete using for loops 
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
```

```python
#checking the ingredients 
df.dtypes
```

```python
#Amending the ingredients 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df ['gross'].astype('int64')

df
```

```python
#Cleaning up the plates 
df['year_adjustment']= df['released'].astype(str).str[0:4]

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Checking for the quality of the dishes 

df['country'].drop_duplicates().sort_values(ascending=False)
df.drop_duplicates()
```

```python
# comparing the dishes (buget & gross)

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Relationship between Budget & Gross Income')
plt.xlabel('Budget')
plt.ylabel('Gross Income')
plt.show
```

```python
# adding new ingredients in using Seaborn

sns. regplot(x='budget', y='gross', data=df, scatter_kws={'color': "red"}, line_kws= {"color":"blue"})
```

```python
# see whether if the ingredients is compatible 

df.corr()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation Indicators')

plt.xlabel ('Movie Attributes')

plt.ylabel ('Movie Attributes')

plt.show()
```

```python
df_numerised = df

for col_name in df_numerised.columns:
    if(df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name]= df_numerised[col_name].astype('category')
        df_numerised[col_name]= df_numerised[col_name].cat.codes
        
df_numerised
```

```python
correlation_mat = df_numerised.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs [(sorted_pairs) > 0.5]

high_corr
```

```python
# Assigning the dishes to its unique catergory using factorise 

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
